```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load
cat = cv.imread('./cat.jpg', flags=1)

# we are using RGB
cat = cv.cvtColor(cat, cv.COLOR_BGR2RGB)
```

```python
# show
plt.imshow(cat)
plt.title('cat')
plt.axis('off')
plt.show()
```

```python
# data shape = (height, width, channel)
cat.shape
```

```python
# black and white
def black():
    black = np.ones((512,512,3), np.uint8) # np.uint8: 0~255
    return black

def white():
    white = np.ones((512,512,3), np.uint8) * 255 # np.uint8: 0~255
    return white

black_n_white = np.concatenate((black(), white()), axis=1)
plt.imshow(black_n_white)
```

```python
# resize
rs_cat = cv.resize(cat,(512,512))
plt.imshow(rs_cat)
```

```python
# crop
plt.imshow(rs_cat[100:400, 100:400])
```

```python
# draw line
white = white()
plt.imshow(cv.line(white, (0,0), (512,512), (255,0,0), 5))
plt.imshow(cv.line(white, (512,0),(0,512), (255,0,0), 5))
```

```python
# draw circle
plt.imshow(cv.circle(white, (256,256), 100, (255,0,0), -1))
```

```python
# put text
plt.imshow(cv.putText(black(), 'OpenCV', (80,200), cv.FONT_HERSHEY_SCRIPT_SIMPLEX, 3, (255,255,255)))
```

```python
# blue canvas
ones = np.ones((512,512), dtype=np.uint8)
fulls = ones * 255
blue = np.stack((ones,ones,fulls), axis=2)
plt.imshow(blue)
```

```python
# image calculation
plt.imshow(cv.add(blue, rs_cat))
```
